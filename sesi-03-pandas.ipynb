# SESI 03 - PANDAS: MANIPULASI DATA

Pandas dalah library tama Python untuk analisis data.
Cara kerjanya mirip Excel, tapi jauh lebih powerful dan isa handle data berukuran besar.

## 1. Import Pandas & Buat Data

```python
import pandas as pd

# Buat data frame - seperti tabel di Excel
data = {
    "nama": ["Andi", "Budi", "Citra", "Deni", "Endra"],
    "nilai": [92, 75, 88, 60, 95],
    "kota": ["Jakarta", "Bandung", "Jakarta", "Surabaya", "Bandung"],
    "lulus": [True, True, True, False, True]
}

df = pd.DataFrame(data)
print(df)
```

## 2. Eksplorasi Data Dasar

```python
print(df.shape)   #jumlah baris & kolom
print(df.columns) #nama kolom
print(df.dtypes)   #tipe data tiap kolom
print(df.describe()) #statistik dasar
```

## 3. Filter Data

Ambil baris tertentu berdasarkan kondisi.

```python
# Ambil siswa yang lulus saja

lulus = df[df["lulus"] == True]
print(lulus)
```

```python
# Ambil siswa dengan nilai di atas 80

nilai_tinggi = df[df["nilai"] > 80]
print(nilai_tinggi)
```

```python
# Ambil siswa dari kota Bandung

Bandung = df[df["kota"] == "Bandung"]
print(Bandung)
```

## 4. Groupby - Agregasi per Kelompok

Groupby dipakai untuk merangkum data per kategori - misalnya rata-rata nilai per kota.

```python
#Rata-rata nilai per Kota
per_kota = df.groupby("kota")["nilai"].mean()
print(per_kota)
```

```python
#Jumlah siswa per Kota
jumlah = df.groupby("kota")["nama"].count()
print(jumlah)
```

## 5. Tambah Kolom Baru & Export to CSV

```python
#Tambah kolom kategori berdasarkan nilai
def kategorikan(nilai):
    if nilai >= 90:
        return "Sangat Baik"
    elif nilai >= 80:
        return "Baik"
    elif nilai >=70:
        return "Cukup"
    else:
        return "Kurang"

df["kategori"] = df["nilai"].apply(kategorikan)
print(df)
```

```python
#Simpan ke file CSV
df.to_csv("hasil_siswa.csv", index=False)
print("File berhasil disimpan!")
```
