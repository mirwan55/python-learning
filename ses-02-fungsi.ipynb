# Sesi 02 - Fungsi (def)

Fungsi adalah kode yang bisa dipanggil berkali-kali
Daripada nulis kode yang sama berulang, kita bungkus dalam fungsi.

## 1. Fungsi Dasar

```python
def sapa(nama):
    print(f"Halo, {nama}! Selamat belajar Python.")

sapa("Andi")
sapa("Budi")
sapa("Cia")
```

## 2. Fungsi dengan Return

Fungsi bisa mengembalikan nilai dengan 'return'
Hasilnya bisa disimpan ke variabel.

```python
def hitung_rata_rata(nilai):
    total = sum(nilai)
    rata = total / len(nilai)
    return rata

nilai_kelas = [85, 92, 78, 90, 88]
hasil = hitung_rata_rata(nilai_kelas)
print(f"Rata-rata kelas: {hasil}")
```

## 3. Fungsi dengan Defaul Parameter

Parameter bisa punya nilai default
Kalau tidak diisi, pakai nilai default-nya

```python
def perkenalan(nama, kota="Jakarta"):
    print(f"Nama saya {nama}, tinggal di {kota}.")

perkenalan("Andi", "Bandung")
perkenalan("Budi") #kota tidak diisi, akan dipakai default "Jakarta"
```

## 4. Fungsi untuk Analisis Data Sederhana

Contoh nyata penggunaan fungsi dalam konteks data analyst.

```python
def analisis(data, label="Data"):
    print(f"=== Analisis {label} ===")
    print(f"Jumlah data : {len(data)}")
    print(f"Nilai min   : {min(data)}")
    print(f"Nilai max   : {max(data)}")
    print(f"Rata-rata   : {sum(data)/len(data):.2f}")

nilai = [85, 92, 78, 90, 88, 70, 95]
analisis(nilai, "Nilai Ujian")
```

## 5. Fungsi di dalam Loop

Fungsi bisa dipanggil berulang di dalam loop
kombinasi ini serinf dipakai saat memproses banyak data.

```python
def kategorikan(nilai):
    if nilai >= 90:
        return "Sangat Baik"
    elif nilai >=80:
        return "Baik"
    elif nilai >=70:
        return "Cukup"
    else:
        return "Kurang"

data_siswa = [
    {"nama": "Andi", "nilai": 92},
    {"nama": "Budi", "nilai": 75},
    {"nama": "Citra", "nilai": 88},
    {"nama": "Deni", "nilai": 60}
]

for siswa in data_siswa:
    kategori = kategorikan(siswa["nilai"])
    print(f"{siswa['nama']}: {siswa['nilai']} → {kategori}")
```
