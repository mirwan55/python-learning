# Sesi 04 - Visualisasi Data

Visualisasi adalah cara menyajikan data dalam bentuk grafik.
Kita akan pakai dua library: Matplotlib (dasar) dan Seaborn (lebih bagus dan estetik).

```python
## 1. Setup Data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data yang akan dilakukan visualisasi

data = {
    "nama": ["Andi", "Budi", "Citra", "Dendi", "Emil"],
    "nilai": [90, 78, 86, 65, 93],
    "kota": ["Jakarta", "Surabaya", "Ambon", "Aceh", "Manado"],
    "kategori": ["Sangat Baik", "Cukup", "Baik", "Kurang", "Sangat Baik"]
}

df = pd.DataFrame(data)
print(df)
```

## 2. Bar Chart Nilai Per Siswa

```python
plt.figure(figsize=(5, 3))
plt.bar(df["nama"], df["nilai"], color="orange")
plt.title("Nilai per Siswa")
plt.xlabel("Nama Siswa")
plt.ylabel("Nilai Siswa")
plt.ylim(0, 100)
plt.show()
```

## 3. Line Chart Tren Nilai

```python
plt.figure(figsize=(5, 3))
plt.plot(df["nama"], df["nilai"], color="steelblue", marker="x", linewidth=3)
plt.title("Tren Nilai Siswa")
plt.xlabel("Nama Siswa")
plt.ylabel("Nilai Siswa")
plt.ylim(0, 100)
plt.grid(True)
plt.show
```

## 4. Pie Chart Distribusi Kategorisasi Siswa

```python
jumlah_kategori = df["kategori"].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(jumlah_kategori, labels=jumlah_kategori.index, autopct="%1.1f%%")
plt.title("Distribusi Kategori Nilai")
plt.show()
```

## 5. Seaborn Bar Chart

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="nama", y="nilai", hue="kategori")
plt.title("Nilai Siswa per Kategori")
plt.xlabel("Nama")
plt.ylabel("Nilai")
plt.show()
```
