```python
# Variables - wadah untuk menyimpan data
nama = 'Mirwan'
umur = 19
tinggi = 180
aktif = True

print(nama)
print(umur)
print(tinggi)
print(aktif)
```

```python
# CEK TIPE DATA dengan type()
nama = 'Mirwan'
umur = 19
tinggi = 180.5
aktif = True

print(type(nama)) #str = teks
print(type(umur)) #int = bilangan bulat
print(type(tinggi)) #float = bilangan desimal
print(type(aktif)) # bool = True/False
```

```python
# OPERASI MATEMATIKA
a = 100
b = 30

print(a + b) #penjumlahan
print(a - b) #pengurangan
print(a * b) #perkalian
print(a / b) #pembagian
print(a // b) #pembagian bulat (buang desimal)
print(a % b) #sisa bagi
print(a ** 2) #pangkat
```

```python
# OPERASI PAD TEKS (STRING)
nama_depan = 'Mirwan'
nama_belakang = 'Belajar'

nama_lengkap = nama_depan + " " + nama_belakang #gabung teks
print(nama_lengkap)
print(len(nama_lengkap)) #hitung jumlah karakter
print(nama_lengkap.upper()) #huruf besar semua
print(nama_lengkap.lower()) #huruf kecil semua
print(nama_lengkap.replace("Mirwan", "Bro")) #ganti kata
```

```python
# LIST - seperti kolom di Excel
nilai = [85, 92, 78, 90, 88]

print(nilai)
print(nilai[0]) #ambil data pertama (index mulai dari 0)
print(nilai[-1]) #ambil data terakhir
print(nilai[1:4]) #ambil index 1 sampai 3
print(len(nilai)) #jumlah data
print(sum(nilai)) #total
print(sum(nilai) / len(nilai)) #rata-rata manual
```

```python
# LIST - manipulasi LIST
buah = ['apel', 'mangga', 'jeruk']
buah.append('pisang') #tambah di akhir
print(buah)

buah.remove('mangga') #hapus by nilai
print(buah)

print('apel' in buah) #cek apakah ada
```

```python
# DICTIONARY - seperti satu baris di tabel excel
pegawai = {
    'nama':'Mirwan',
    'umur':19,
    'jabatan':'Data Analyst',
    'aktif':True
}

print(pegawai['nama']) #ambil by key
print(pegawai['jabatan']) #ambil by key

pegawai['kota'] = 'Jakarta' #tambah data baru
print(pegawai)
```

```python
# KONDISI IF/ELSE
nilai = 85

if nilai >= 90:
    print("A - sangat baik")
elif nilai >= 80:
    print("B - Baik")
elif nilai >= 70:
    print("C - Cukup")
else:
    print("D - Kurang")
```

```python
# Menyimpan ke dalam variabel
nilai = 90

if nilai >= 90:
    kategori = "Sangat Baik"
elif nilai >= 80:
    kategori = "Baik"
elif nilai >= 70:
    kategori = "Cukup"
else:
    kategori = "Kurang"

print(f"Nilai {nilai} masuk kategori: {kategori}")
```

```python
# Materi LOOP(FOR)
nilai = [85, 92, 78, 90, 88]

for n in nilai:
    print(n)
```

```python
# LOOP dengan kondisi
nilai = [85, 92, 78, 90, 88]

for n in nilai:
    if n >= 90:
        print(f"{n} → Sangat Baik")
    elif n >= 80:
        print(f"{n} → Baik")
    elif n >= 70:
        print(f"{n} → Cukup")
    else:
        print(f"{n} → Kurang")
```
